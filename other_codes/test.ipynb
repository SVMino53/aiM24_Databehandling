```python
my_ints = [3, 5, 5, 7, 8, 12]

ints_series = pd.Series(my_ints)

print(ints_series)
```

```python
print("'ints_series' data:")
print(f"  - Minsta värdet:\t{ints_series.min()}")
print(f"  - Största värdet:\t{ints_series.max()}")
print(f"  - Medelvärdet:\t{ints_series.mean()}")
print(f"  - Medianvärdet:\t{ints_series.median()}")
```

```python
students = ["Isak", "Astrid", "Ralph", "Viktoria", "Emma", "Aldin"]
ages = [26, 22, 29, 22, 25, 32]
grades = [4, 3, 3, 2, 4, 3]

aim_class_df = pd.DataFrame({"name": students,
                             "age": ages,
                             "grade": grades})
```
