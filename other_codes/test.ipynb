```python
import random


class Node:
    def __init__(self, name : str, parent : "Node" = None) -> None:
        self.name = name
        self.parent = parent
        if isinstance(parent, Node):
            parent.__add_child(self)
        self.children : list[Node] = []

    def __add_child(self, child : "Node") -> None:
        self.children.append(child)
    
    def get_child(self, index : int) -> "Node":
        if len(self.children) > 0 and -len(self.children) <= index < len(self.children):
            return self.children[index]
        return None
    
    def get_children(self) -> list["Node"]:
        return self.children
    
    def get_level_nodes(self, level : int) -> list["Node"]:
        if level >= self.get_depth():
            return []
        elif level == 0:
            return [self]
        elif level == 1:
            return self.children
        else:
            return [n for c in self.children for n in c.get_level_nodes(level - 1)]
    
    def get_all_children(self) -> list["Node"]:
        all_children = list(self.children)
        for child in self.children:
            all_children += child.get_all_children()
        return all_children
    
    def get_level_width(self, level : int) -> int:
        return len(self.get_level_nodes(level))

    def get_max_width(self) -> int:
        max_width = 1
        for lvl in range(1, self.get_depth()):
            cur_width = self.get_level_width(lvl)
            if cur_width > max_width:
                max_width = cur_width
        return max_width

    def get_depth(self) -> int:
        deepest = 1
        for child in self.children:
            cur_depth = child.get_depth() + 1
            if cur_depth > deepest:
                deepest = cur_depth
        return deepest

    def __str__(self) -> str:
        cc = {"-": "\u2504",
              "/": "\u250c",
              "\\": "\u2510",
              ">": "\u251c",
              "<": "\u2524",
              "T": "\u252c",
              "^": "\u2534",
              "+": "\u253c"}
        rows = []
        if isinstance(self.parent, Node):
            name_len = len(self.name)
            par_name_len = len(self.parent.name)
            name_len_dif = abs(name_len - par_name_len)
            if name_len < par_name_len:
                rows += [self.parent.name, f"{" "*(par_name_len//2 - 1)}..", f"{" "*(name_len_dif//2)}{self.name}"]
            else:
                rows += [f"{" "*(name_len_dif//2)}{self.parent.name}", f"{" "*(name_len//2 - 1)}..", self.name]
        else:
            rows.append(self.name)
        for _ in range(self.get_depth()):
            rows += ["---", "nodes"]


        return "\n".join(rows)
    
    def __repr__(self):
        return f"Node({self.name})"
```

```python
root = Node('root_node')
for i in range(20):
    parent = root
    while len(parent.children) > 0:
        if random.random() < 0.6:
            parent = random.choice(parent.children)
        else:
            break
    Node(f"n{i}", parent)
```
