````python
# functions

def set_headers(data_frame : pd.DataFrame, headers : list):
    data_frame.rename(columns={oldk: newk for oldk, newk in zip(data_frame.keys(), headers)}, inplace=True)

````

````python
kommuner_män_df = pd.read_excel(path, "Män", header=6)
kommuner_kvinnor_df = pd.read_excel(path, "Kvinnor", header=6)
````

````python
# a)

new_headers = ["Rang 2020", "Rang 2019", "Kommun", "Folkmängd 2020", "Folkmängd 2019", "Förändring", "Kön"]
set_headers(kommuner_män_df, new_headers)
set_headers(kommuner_kvinnor_df, new_headers)

kommuner_män_df["Kön"] = "Man"
kommuner_kvinnor_df["Kön"] = "Kvinna"

kommuner_kvinnor_df
````

````python
# b)

kommuner_båda_df = pd.concat([kommuner_män_df, kommuner_kvinnor_df])
kommuner_båda_df.pop("Rang 2020")
kommuner_båda_df.pop("Rang 2019")
kommuner_båda_df.set_index("Kommun", inplace=True)
kommuner_båda_df
````

````python
# c)

kommuner_df.pop("Rang 2020")
kommuner_df.pop("Rang 2019")
kommuner_df.set_index("Kommun", inplace=True)
kommuner_df.sort_values("Kommun", inplace=True)
kommuner_df.rename(columns={"Folkmängd 2020": "Total Pop 2020",
                            "Folkmängd 2019": "Total Pop 2019",
                            "Förändring": "Total förändring (%)"}, inplace=True)
kommuner_df
````

````python
# d)

kommuner_kön_o_total_df = kommuner_båda_df.merge(kommuner_df, "outer", ["Kommun"])
kommuner_kön_o_total_df
````

````python
# e)

kommuner_kön_o_total_df.sort_values("Total Pop 2020", ascending=False, inplace=True)
topp10_kommuner_df = kommuner_kön_o_total_df.head(20)
botten10_kommuner_df = kommuner_kön_o_total_df.tail(20)

ax = sns.barplot(topp10_kommuner_df, x="Folkmängd 2020", y="Kommun", hue="Kön", hue_order=["Man", "Kvinna"])
ax.bar_label(ax.containers[0], color="white", label_type="center")
ax.bar_label(ax.containers[1], color="white", label_type="center")
ax.set_title("Topp 10 Kommuner")
plt.show()

ax = sns.barplot(botten10_kommuner_df, x="Folkmängd 2020", y="Kommun", hue="Kön", hue_order=["Man", "Kvinna"])
ax.bar_label(ax.containers[0], color="white", label_type="center")
ax.bar_label(ax.containers[1], color="white", label_type="center")
ax.set_title("Nedre 10 Kommuner")
plt.show()
````

````python
# f)

antal_män = kommuner_män_df["Folkmängd 2020"].sum()
antal_kvinnor = kommuner_kvinnor_df["Folkmängd 2020"].sum()
fig, ax = plt.subplots()
ax.pie([antal_män, antal_kvinnor], labels=["Män", "Kvinnor"], startangle=90, autopct=lambda pct: f"{pct:.2f}%\n({round(pct/100*(antal_män + antal_kvinnor))} st)")
ax.set_title("Antal män och kvinnor i Sverige")

plt.show()
````

````python
kommuner_män_df.set_index("Kommun", inplace=True)
kommuner_kvinnor_df.set_index("Kommun", inplace=True)
````

````python
# g)

topp5_kommuner_könsskilnad = {"Kommun": [], "Skillnad": []}
for kommun in kommuner_män_df.index:
    antal_män = kommuner_män_df.loc[kommun, "Folkmängd 2020"]
    antal_kvinnor = kommuner_kvinnor_df.loc[kommun, "Folkmängd 2020"]
    pct_skillnad = abs(antal_män - antal_kvinnor)/max(antal_män, antal_kvinnor)*100
    if topp5_kommuner_könsskilnad["Kommun"] == []:
        topp5_kommuner_könsskilnad["Kommun"].append(kommun)
        topp5_kommuner_könsskilnad["Skillnad"].append(pct_skillnad)
    for i, jämf_skillnad in enumerate(topp5_kommuner_könsskilnad["Skillnad"]):
        if pct_skillnad > jämf_skillnad:
            topp5_kommuner_könsskilnad["Kommun"].insert(i, kommun)
            topp5_kommuner_könsskilnad["Skillnad"].insert(i, pct_skillnad)
            break
    if len(topp5_kommuner_könsskilnad["Kommun"]) > 5:
        topp5_kommuner_könsskilnad["Kommun"].pop()
        topp5_kommuner_könsskilnad["Skillnad"].pop()

topp5_kommuner_könsskilnad_df = pd.DataFrame(topp5_kommuner_könsskilnad)
topp5_kommuner_könsskilnad_df.rename(columns={"Skillnad": "Könsskillnad (%)"}, inplace=True)
ax = sns.barplot(topp5_kommuner_könsskilnad_df, y="Kommun", x="Könsskillnad (%)")
ax.bar_label(ax.containers[0], color="white", label_type="center", fmt=lambda x: f"{x:.2f}%")

plt.show()
````

````python
kommuner_män_df[["Folkmängd 2020","Folkmängd 2019"]]
````

````python
# g) optimal

kommuner_män_o_kvinnor_df = pd.merge(kommuner_män_df["Folkmängd 2020"], kommuner_kvinnor_df["Folkmängd 2020"], on="Kommun")
kommuner_män_o_kvinnor_df.rename(columns={"Folkmängd 2020_x": "Folkmängd 2020 Män",
                                          "Folkmängd 2020_y": "Folkmängd 2020 Kvinnor"}, inplace=True)
kommuner_män_o_kvinnor_df["Könsskillnad (Antal)"] = abs(kommuner_män_o_kvinnor_df["Folkmängd 2020 Män"] - kommuner_män_o_kvinnor_df["Folkmängd 2020 Kvinnor"])
kommuner_män_o_kvinnor_df["Könsskillnad (%)"] = kommuner_män_o_kvinnor_df["Könsskillnad (Antal)"]/(kommuner_män_o_kvinnor_df["Folkmängd 2020 Män"] + kommuner_män_o_kvinnor_df["Folkmängd 2020 Kvinnor"])*100
kommuner_män_o_kvinnor_df.sort_values(by="Könsskillnad (%)", ascending=False, inplace=True)
topp5_kommuner_könsskillnad_df = kommuner_män_o_kvinnor_df.head()

ax = sns.barplot(topp5_kommuner_könsskillnad_df, y="Kommun", x="Könsskillnad (%)")
ax.bar_label(ax.containers[0], color="white", label_type="center", fmt=lambda x: f"{x:.2f}%")
plt.show()
````

````python
# h)

kommuner_df["Total förändring (antal)"] = kommuner_df["Total Pop 2020"] - kommuner_df["Total Pop 2019"]
kommuner_df.sort_values("Total förändring (antal)", ascending=False, inplace=True)

topp5_kommuner_förändring_df = kommuner_df.head()
ax = sns.barplot(topp5_kommuner_förändring_df, x="Kommun", y="Total förändring (antal)")
ax.set_title("Topp 5 Kommuner\nÖkning i Antal Invånare 2019-2020")
plt.show()
````
